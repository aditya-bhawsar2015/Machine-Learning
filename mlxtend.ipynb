```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
fp_df = pd.read_csv(r'D:\workspace\MachineLearning\Datasets\Datasets\Faceplate.csv', index_col=0)
fp_df = fp_df.astype(bool)
fp_df
```

```python
itemsets = apriori(fp_df, min_support=0.2, use_colnames=True)
itemsets
```

```python
itemsets = apriori(fp_df, min_support=0.3, use_colnames=True)
rules = association_rules(itemsets, metric='confidence', min_threshold=0.6)
rules      # rules having confidence > 0.6 
```

```python
rules = rules[['antecedents','consequents','support','confidence','lift']]
rules.sort_values('lift', ascending=False)
```

#### Groceries

The columns in the above dataset and the entries are not of equal lengths hence we need to code it in python code

Can also be done using the sklearn CountVectorizer also

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
groceries = []
with open('D:\workspace\MachineLearning\Datasets\Datasets\Groceries.csv','r') as f:groceries = f.read()
```

```python
groceries = groceries.split('\n')
```

```python
groceries_list =  []
for i in groceries:
    groceries_list.append(i.split(','))
```

```python
te = TransactionEncoder()
te_ary = te.fit(groceries_list).transform(groceries_list)
grocery_df = pd.DataFrame(te_ary, columns=te.columns_)
grocery_df
```

```python
itemsets = apriori(grocery_df, min_support=0.005, use_colnames=True)
rules = association_rules(itemsets, metric='confidence', min_threshold=0.6)
rules = rules[['antecedents','consequents','support','confidence','lift']]
rel_rules = rules[rules['lift']>1]
rel_rules.sort_values('confidence', ascending=False)
rules
```
