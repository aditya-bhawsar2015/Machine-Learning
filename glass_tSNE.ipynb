```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
import seaborn as sns
```

```python
glass = pd.read_csv(r'D:\workspace\MachineLearning\Cases\Glass Identification\Glass.csv')
X,y = glass.drop('Type', axis=1), glass['Type']
```

```python
model = TSNE(random_state=25).set_output(transform='pandas')
tsne_features = model.fit_transform(X)
```

```python
tsne_features['Type'] = y
tsne_features.columns
```

```python
plt.figure(figsize=(18,12))
sns.scatterplot(data=tsne_features, x='tsne0', y='tsne1', hue='Type')
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
prcomp = PCA(n_components=2).set_output(transform='pandas')
pca_transform = prcomp.fit_transform(X)

pca_transform['Type'] = y
```

```python
plt.figure(figsize=(18,12))
sns.scatterplot(data=pca_transform, x='pca0', y='pca1', hue='Type')
plt.show()
```
