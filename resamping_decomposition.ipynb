```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\workspace\MachineLearning\Datasets\Datasets\monthly-milk-production-pounds-p.csv', index_col=0)
df.index = pd.DatetimeIndex(df.index)
```

```python
df.plot()
plt.title('Monthly Milk Production')
plt.xlabel('Months')
plt.show()
```

```python
downsampled = df.resample('Q').sum()
downsampled.index.rename('Quater', inplace=True)
downsampled.plot()
plt.title('Quaterly Milk Production')
plt.xlabel('Quaters')
plt.show()
```

#### Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
series = df['Milk']
result = seasonal_decompose(series, model='additive', period=12)
result.plot()
plt.title('Additive Decompose')
plt.show()
```

```python
series = df['Milk']
result = seasonal_decompose(series, model='multiplicative', period=12)
result.plot()
plt.title('multiplicative Decompose')
plt.show()
```

```python
series = downsampled['Milk']
result = seasonal_decompose(series, model='additive', period=4)
result.plot()
plt.title('Additive Decompose for Quaterly data')
plt.show()
```
